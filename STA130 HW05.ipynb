To reject the null hypothesis, we need enough evidence from our observed data. For that we need to know how likely null hypothesis is actually true. This is why we "imagine a world where the null hypothesis is true" so that we can observe data as extreme as or even more extreme than the data we collected. This is done by generating multiple simulations under the null hypothesis. These data are compared to our observed data and the p-value is extracted. This experiment is similar to Fisher’s original experiment with Dr. Bristol, where she had to identify the pouring order of milk and tea. 

Population: The original population was Dr. Bristol, a single individual, whereas this experiment deals with 80 random students from STA130.
Sample Size: In the original experiment, Dr. Bristol was presented with 8 cups of tea. Here, we have a larger sample size of 80 students.
Personalized Perception vs. General Guessing: Fisher’s original experiment tested an individual’s perception, while this test generalizes perception for a group of students, which might be more abstract.

Null Hypothesis (H₀): The students are guessing randomly, meaning the probability of correctly identifying whether milk or tea was poured first is 0.5 (the same as a fair coin toss).
Formally: H_0: p = 0.5 
Informal Interpretation: This means we assume students are just guessing, and about half of them would get the correct answer by chance.

Alternative Hypothesis (H₁):
The proportion of students who can correctly identify the pouring order is greater than 0.5. This implies students have some ability to detect the correct order.
Formally: H_1: p > 0.5 

Quantitative Analysis:

To assess the validity of the null hypothesis, we conduct a simulation-based approach. We will simulate a binomial distribution under the null hypothesis to estimate the probability of observing 49 or more correct identifications out of 80 trials.

Explanation of the Code

1.	Parameters: We define the sample size (n), the observed number of correct identifications (x), the null hypothesis probability (p0), and the number of simulations (num_simulations).
2.	Reproducibility: The random seed is set to ensure that the results can be reproduced.
3.	Simulation: We generate 10,000 simulated samples of successes based on a binomial distribution with parameters  n  and  p_0 .
4.	P-value Calculation: We calculate the p-value by determining the proportion of simulated successes that are greater than or equal to the observed number of correct identifications

```python
import numpy as np

# Parameters
n = 80  # sample size
x = 49  # observed number of correct identifications
p0 = 0.5  # null hypothesis probability
num_simulations = 10000  # number of simulations

# Seed for reproducibility
np.random.seed(42)

# Run the simulations
simulated_successes = np.random.binomial(n, p0, num_simulations)

# Calculate p-value
p_value = np.mean(simulated_successes >= x)

p_value
```

Since our p_value is in between 0.01 < p-value <= 0.05(moderate evidence), we do not have enough evidence to reject the null hypothesis. P_value also indicates that their performance is consistent with random guessing.
